```python
%load_ext autoreload
%autoreload 2
%matplotlib inline
import ipywidgets as w
import matplotlib.pyplot as plt
import numpy as np
import netCDF4
plt.rcParams.update({'font.size': 16})
from matplotlib.ticker import (MultipleLocator, FormatStrFormatter,
                               AutoMinorLocator)
from scipy.stats import skew
from typhon.retrieval.qrnn import set_backend, QRNN
set_backend("pytorch")
import stats as S
```

## INPUT PARAMETERS 

```python
depth     = 4
width     = 128
quantiles = np.array([0.002, 0.03, 0.16, 0.5, 0.84, 0.97, 0.998])
batchSize = 128

target = 'MWI-15'


#inChannels = np.array(['I1V', 'I2V', 'I3V', 'I5V' , 'I6V', 'I7V', 'I8V', 'I9V', 'I10V', 'I11V', 'I11H'])
inChannels = np.array([target, 'I5V', 'I6V', 'I7V', 'I8V', 'I9V', 'I10V', 'I11V'])

i183, = np.argwhere(inChannels == target)[0]
```

### read test data

```python
from ici_mwi import iciData

data = iciData("TB_ICI_mwi_test.nc", 
               inChannels, target, 
               batch_size = batchSize)  

x_noise = data.add_noise(data.x, data.index)
```

```python
file = 'qrnn_ici_mwi_%s_%s_%s_single.nc'%(depth, width, target)
print (file)
qrnn = QRNN.load(file)
```

#### predict the expectation value

```python
y_pre, y_prior, y0, y, y_pos_mean = S.predict(data, qrnn, add_noise = True)
```

```python
binstep = 0.5
bins = np.arange(-20, 15, binstep)
iq = np.argwhere(quantiles == 0.5)[0,0]
```

```python
hist_noise  = S.calculate_pdf(y, y0, bins)

hist_pre    = S.calculate_pdf(y_pre[:, iq], y0, bins)

hist_prior  = S.calculate_pdf(y_prior[:, i183], y0, bins)

hist_pos_mean    = S.calculate_pdf(y_pre[:, iq], y0, bins)

im = np.abs(y_pos_mean - y_prior[:, i183]) <= 5.0
hist_pos_mean_5  = S.calculate_pdf(y_pre[:, iq], y0, bins, im)

im = np.abs(y_pos_mean - y_prior[:, i183]) <= 10.0
hist_pos_mean_10 = S.calculate_pdf(y_pre[:, iq], y0, bins, im)

im = np.abs(y_pos_mean - y_prior[:, i183]) <= 15.0
hist_pos_mean_15 = S.calculate_pdf(y_pre[:, iq], y0, bins, im)


```

```python
fig, ax = plt.subplots(1,1, figsize = [8,8])
center = (bins[:-1] + bins[1:])/2
ax.plot(center, hist_noise[0])
ax.plot(center, hist_prior[0])
ax.plot(center, hist_pre[0])
#ax.plot(center, hist_pos_mean[0])
#ax.plot(center, hist_pos_mean_5[0])
#ax.plot(center, hist_pos_mean_10[0])
ax.plot(center, hist_pos_mean_15[0])

ax.set_yscale('log')
ax.set_ylabel('Occurence frequency [#/K]')
ax.set_xlabel('Deviation to noise free clear-sky [K]')
ax.xaxis.set_minor_locator(MultipleLocator(1))
ax.grid(which = 'both', alpha = 0.2)


ax.legend(["Noise", "Uncorrected", "Predicted (all)", "Predicted(5K)", "Predicted(15K)" ], prop={'size': 10})

ax.set_title("Channel :%s"%target)

#fig.savefig('Figures/ICI_%s_mwi_single.eps'%str(target),  format='eps')
fig.savefig('Figures/ICI_%s_mwi_single'%str(target))
```
